```python
class CoffeeMachine:
    running=False
    
    def __init__(self,water,milk,coffee_beans,cups,money):
        self.water=water
        self.milk=milk
        self.coffee_beans=coffee_beans
        self.cups=cups
        self.money=money
        
        if not CoffeeMachine.running:
            self.start()
        
    def start(self):
        self.running=True
        self.action=input("Enter the action: \nBuy-want to buy a coffee\nFill-get the stocks\nTake-give the money\nReport-get the amount remaining\nOff-close the machine):\n")
        print()
        action_choices = {"Buy":self.buy,"Fill":self.fill, "Take":self.take, "Off":exit, "Report":self.status}
        if self.action in action_choices:
            action_choices[self.action]()
        else:
            print("Thank you for visiting!\nHope to see you again\n")
            exit()
        
    def return_to_menu(self):
        print()
        self.start()
    
    def available_check(self):
        self.not_available=""
        if self.water-self.reduced[0]<0:
            self.not_available="water"
        elif self.milk-self.reduced[1]<0:
            self.not_available="milk"
        elif self.coffee_beans-self.reduced[2]<0:
            self.not_available="coffee beans"
        elif self.cups-self.reduced[3]<0:
            self.not_available="disposable cups"
        
        if self.not_available != "":
            print(f"Sorry, not enough{self.not_available}!")
            return False
        else:
            print("I have enough resources, making you a coffee!")
            return True
        
    def deduct_supplies(self):
        self.water -= self.reduced[0]
        self.milk -= self.reduced[1]
        self.coffee_beans -= self.reduced[2]
        self.cups -= self.reduced[3]
        self.money += self.reduced[4]
        
    def buy(self):
        self.choice = input("What do you want to buy?\n1 - Espresso\n2 - Latte\n3 - Cappuccino\nback - to main menu:\n")
        self.cost=[3.50,5.60,4.89]
        print("\nCoffee\t\tCost\nExpresso\t$3.50\nLatte\t\t$5.60\nCappuccino\t$4.89")
        if self.choice == '1':
            y=int(input("Enter the number of quarters: "))
            z=int(input("Enter the number of dimes: "))
            w=int(input("Enter the number of nickles:"))
            p=int(input("Enter the number of pennies: "))
            x=0.25*y+0.10*z+0.05*w+0.01*p
            if x>self.cost[0]:
                print("\nThe change: ",x-self.cost[0],"is transferred to your account\n")
            elif x<self.cost[0]:
                y1=float(input("Enter the remaining money: "))
                if (x+y1)==self.cost[0]:
                    print("\nThank you! Your Expresso will be ready in a minute.")
            self.reduced = [250, 0, 16, 1, 3.50] 
            if self.available_check():
                print("\nEnjoy your Expresso\n")
                self.deduct_supplies()

        elif self.choice == '2':
            y=int(input("Enter the number of quarters: "))
            z=int(input("Enter the number of dimes: "))
            w=int(input("Enter the number of nickles:"))
            p=int(input("Enter the number of pennies: "))
            x=0.25*y+0.10*z+0.05*w+0.01*p
            if x>self.cost[1]:
                print("\nThe change: ",x-self.cost[1],"is transferred to your account\n")
            if x<self.cost[1]:
                y1=float(input("Enter the remaining money: "))
                if (x+y1)==self.cost[1]:
                    print("\nThank you! Your Latte will be ready in a minute.")
            self.reduced = [350, 75, 20, 1, 5.60]
            if self.available_check():
                print("\nEnjoy your Latte\n")
                self.deduct_supplies()

        elif self.choice == "3":
            y=int(input("Enter the number of quarters: "))
            z=int(input("Enter the number of dimes: "))
            w=int(input("Enter the number of nickles:"))
            p=int(input("Enter the number of pennies: "))
            x=0.25*y+0.10*z+0.05*w+0.01*p
            if x>self.cost[2]:
                print("\nThe change: ",x-self.cost[2],"is transferred to your account\n")
            if x<self.cost[2]:
                y1=float(input("Enter the remaining money: "))
                if (x+y1)==self.cost[2]:
                    print("\nThank you! Your Cappuccino will be ready in a minute.")
            self.reduced = [200, 100, 12, 1, 4.89]
            if self.available_check():
                print("\nEnjoy your Cappuccino\n")
                self.deduct_supplies()

        elif self.choice == "back": 
            self.return_to_menu()

        self.return_to_menu()
    
    def fill(self): 
        self.water += int(input("Enter the ml of water:\n"))
        self.milk += int(input("Enter the ml of milk:\n"))
        self.coffee_beans += int(input("Enter the grams of coffee beans add:\n"))
        self.cups += int(input("Enter the number of disposable cups:\n"))
        self.return_to_menu()

    def take(self):
        print(f"I gave you ${self.money}")
        self.money -= self.money
        self.return_to_menu()

    def status(self): 
        print(f"The coffee machine has:")
        print(f"{self.water} of water")
        print(f"{self.milk} of milk")
        print(f"{self.coffee_beans} of coffee beans")
        print(f"{self.cups} of disposable cups")
        print(f"${self.money} of money")
        self.return_to_menu()

CoffeeMachine(400, 540, 120, 9, 550)
```
